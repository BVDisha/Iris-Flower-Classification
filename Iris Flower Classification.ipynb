```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
print(x_test.head())
```

```python
import numpy as np
x = np.linspace(-6, 6, num = 1000)
plt.figure(figsize = (12,8))
plt.plot(x, 1 / (1 + np.exp(-x))); # Sigmoid Function 
plt.title("Sigmoid Function");
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
model_1 = KNeighborsClassifier(n_neighbors=5)
model_1.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_2=RandomForestClassifier(n_estimators=5)
model_2.fit(x_train, y_train)
```

```python
from sklearn.svm import SVC 
model_3=SVC(kernel='linear')
model_3.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
print(predictions)
```

```python
KNN_predictions = model_1.predict(x_test)
print(KNN_predictions)
```

```python
RF_predictions = model_2.predict(x_test) 
print(RF_predictions)
```

```python
SVM_predictions = model_3.predict(x_test)
print(SVM_predictions)
```

```python
p = pd.DataFrame ([[1,3,5.1,2.4]], columns = ["sepal length", "sepal width","petal length","petal width"])
predictions = model.predict(p) 
print(predictions)
```

```python
print("Accuracy of Logestic regression", model.score(x_test,y_test)*100)
print("Accuracy of knn",model_1.score(x_test,y_test)*100)
print("Accuracy of Random Forest",model_2.score(x_test,y_test)*100)
print("Accuacy of SVM", model_3.score(x_test,y_test)*100)
```
